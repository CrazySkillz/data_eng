```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine

engine = sqlalchemy.create_engine('postgresql://root:@localhost:3305/campaigns')
```

```python
df1 = pd.read_csv('conversions_1.csv')
df1.head()
```

```python
df2 = pd.read_csv('clicks_1.csv')
df2.head()
```

```python
new_df = pd.merge(df1, df2, on='click_id')
new_df = new_df[['campaign_id', 'banner_id','revenue']]
final_df = new_df.sort_values(['campaign_id', 'revenue'], ascending=[True, False]).groupby('campaign_id').head(10)
final_df['banner_pic'] = 'image' 
final_df['banner_pic'] = 'image_'+ final_df['banner_id'].astype(str) + '.png'
final_df
```

```python
final_df.to_sql(
    name='campaigns',
    con=engine,
    index=False
)
```
